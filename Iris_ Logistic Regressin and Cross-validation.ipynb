```python
Isetosa = np.array_split(np.random.permutation(len(setosa)), 5)
Ivirginica = np.array_split(np.random.permutation(len(virginica)), 5)
Iversicolor = np.array_split(np.random.permutation(len(versicolor)), 5)
score = np.zeros(5)
theta = np.zeros(5, dtype=object)
C = np.zeros(5, dtype=object)
for k in range(5):
    Itrainsetosa = np.concatenate((Isetosa[0-k], Isetosa[1-k], Isetosa[2-k], Isetosa[3-k]))
    Itestsetosa = Isetosa[k]
    Itrainvirginica = np.concatenate((Ivirginica[0-k], Ivirginica[1-k], Ivirginica[2-k], Ivirginica[3-k]))
    Itestvirginica = Ivirginica[k]
    Itrainversicolor = np.concatenate((Iversicolor[0-k], Iversicolor[1-k], Iversicolor[2-k], Iversicolor[3-k]))
    Itestversicolor = Iversicolor[k]
    A = np.vstack((setosa[Itrainsetosa], virginica[Itrainvirginica], versicolor[Itrainversicolor]))
    Atest = np.vstack((setosa[Itestsetosa], virginica[Itestvirginica], versicolor[Itestversicolor]))
    ytest = np.concatenate((np.zeros(len(Itestsetosa)), np.ones(len(Itestvirginica)), np.zeros(len(Itestversicolor))))
    y = np.concatenate((np.zeros(len(Itrainsetosa)), np.ones(len(Itrainvirginica)), np.zeros(len(Itrainversicolor))))
    coeff = LogisticRegression().fit(A,y)
    theta[k] = coeff
    score[k] = coeff.score(Atest, ytest)
    C[k] = confusion_matrix(ytest, coeff.predict(Atest))

```
