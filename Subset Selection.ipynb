```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
def flatten(l):
    newl = []
    for i in l:
        if type(i) == list:
            for j in i:
                newl.append(j)
        else:
            newl.append(i)
    return newl
```

```python
def get_binary_repr(x, n):
    l = '0' + str(len(format(n, 'b'))) + 'b'
    return format(x, l)
```

```python
def get_subset(pa):
    p = len(pa)
    subset = {}
    for i in range(p+1):
        subset[i] = []
        
    n = 2**p - 1
    
    for i in range(2**p):
        subset_ = []
        x = get_binary_repr(i, n)
        for i in range(len(x)):
            if x[i] == '1':                
                subset_.append(pa[i])
        subset[len(subset_)].append(subset_)
        
    return subset            
```

```python
def compute_coeff(A,y):
    return np.linalg.inv(A.T @ A) @ A.T @ y
```

```python
credit = pd.read_csv('./datasets/Credit.csv', sep=',')
credit['Bgender'] = credit['Gender'].map({' Male': 1, 'Female': 0})
credit['Bstudent'] = credit['Student'].map({'No': 0, 'Yes': 1})
credit['isAsian'] = credit['Ethnicity'].replace(['Asian', 'Caucasian', 'African American'], [1, 0, 0])
credit['isCaucasian'] = credit['Ethnicity'].replace(['Asian', 'Caucasian', 'African American'], [0, 1, 0])
credit['isMarried'] = credit['Married'].map({'Yes': 1, 'No': 0})
```

```python
pa = ['Income', 'Limit', 'Rating', 'Cards', 'Age', 'Education','Bgender',
       'Bstudent', ['isAsian', 'isCaucasian'], 'isMarried']
subset = get_subset(pa)
```

```python
count = 0
for pa in subset:
    for i in subset[pa]:
        count += 1
        if i != 0:
            y = np.array(credit['Balance'])
            p = flatten(i)
            A = np.column_stack((np.ones(len(y)), credit[p]))
    
            theta = compute_coeff(A, y)
            print(theta)
        else:
            print(credit['Balance'].mean())
```
